```python
import pandas as pd
from pytrends.request import TrendReq

pytrend = TrendReq()
```

```python
pytrends = TrendReq(hl='pl-PL', tz=60)
```

```python
pytrends.build_payload(kw_list=['it job'])

df = pytrends.interest_by_region()
df.sort_values(by=['it job'], ascending=False)
```

According to Google Trends, values ​​are calculated on a scale of 0 to 100, where 100 is the most popular position as part of the total number of searches for that location, a value of 50 indicates a position. position less than twice as common. A value of 0 indicates that the location does not have enough data for that term. Source → Google Trends.



```python
df.reset_index().plot(x='geoName', y='it job', figsize=(120,10), kind='bar')
```

```python
pytrends.build_payload(kw_list=['jobs'])
related_queries = pytrends.related_queries()
related_queries.values()
```

```python
related_topics = pytrends.related_topics()
related_topics.values()
```

```python
kw_list = ['no fluff jobs', 'just join it', 'pracujpl', 'bulldogjob', 'it job']

pytrend.build_payload(kw_list, timeframe = 'today 1-m')
region_df = pytrend.interest_by_region()
region_df = region_df[(region_df != 0).all(1)]

region_df.dropna(how = 'all', axis = 0, inplace = True)

region_df.plot(figsize = (20,12), y = kw_list, kind = 'bar')
```

```python
# historical interest

historical_df = pytrend.get_historical_interest(kw_list, year_start=2022, month_start=1, day_start=1, hour_start=0, year_end=2022, month_end=8, day_end=1, hour_end=0, cat=0, geo='', gprop='', sleep=0)
historical_df.plot(figsize=(20, 12))
historical_df.plot(subplots=True, figsize=(20,12))
```
